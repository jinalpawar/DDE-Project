```python
ess = pd.read_csv("ESS11e04_1/ESS11e04_1.csv")
ess.head()
```

```python
ess_cc = pd.read_csv("ess_country_codes.csv")
ess_cc.head()
```

```python
ess["cntryname"] = ess["cntry"].apply(lambda x: ess_cc[ess_cc["Value "] == x]["Category "].values[0])
```

```python
ess["sclmeet_wg"] = ess["sclmeet"] * ess["anweight"]
ess["rlgblg_wg"] = ess["rlgblg"] * ess["anweight"]
ess["rlgdgr_wg"] = ess["rlgdgr"] * ess["anweight"]
```

```python
df = ess[["cntry", "sclmeet", "rlgblg", "rlgdgr", "sclmeet_wg", "rlgblg_wg", "rlgdgr_wg"]].groupby(by = "cntry", as_index=False).mean()
df.head()
```

```python
df.to_csv("ess_country_level_means.csv")
```
